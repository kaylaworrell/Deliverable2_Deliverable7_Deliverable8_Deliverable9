Based on the Boston_Housing.CSV dataset, it appears that most owner-occupied homes are purchased in the $200,000 range. In addition, it appears that crime rate is much higher around homes that cost less than $200,000.\

```python
import pandas as pd
df= pd.read_csv('/content/boston_housing.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (9,6))
plt.hist(df['MEDV'], color = 'tab:orange')
plt.title('Histogram of Boston Housing Prices', fontsize = 16)
plt.xlabel('Median Value of Owner-Occupied Homes in $10,000\'s', fontsize = 16)
plt.xticks(fontsize = 16)
plt.ylabel('Number of Houses', fontsize = 16)
plt.show()
```

```python
plt.figure(figsize = (9,6))
plt.scatter(x = df['MEDV'], y = df['CRIM'], color = 'Tab:blue')
plt.xlabel('House Price', fontsize = 16)
plt.ylabel('Crime Rate', fontsize = 16)
plt.show()
```
